# Pie Chart

## Import Library for plotting

```python
# import the required libraries
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

# do not display warnings in notebook 
import warnings
warnings.filterwarnings('ignore')
```

Pandas is a great library when it comes to handling a large amount of data. Matplotlib is a visualization library.

## Load Iris dataset

```python
from sklearn import datasets

dataset = datasets.load_iris()
```

We store our data and target into separate pandas DataFrame and perform operations on it.

```python
data = pd.DataFrame(dataset.data)
target = pd.DataFrame(dataset.target)
```

Count the instances in each class using the following functions.

```python
counts = target.count(axis=1).groupby(target[0]).count()
```

```python
labels = 'Setosa', 'Versicolour', 'Virginica'
```

## Plotting

Here we can see what percentage of instance belongs to which class.

```python
fig1, ax1 = plt.subplots(2, 2, figsize=(9, 9))

#labelling pie
fig1.suptitle('Distribution of classes: A pie',
              fontsize=14, fontweight='bold')

# normal plot
ax1[0,0].pie(counts, labels=labels,
             autopct='%1.1f%%')
ax1[0, 0].set_title('Normal Plot')


#plot exploded, default startangle = 0
explode=(0,0,0.1)
ax1[0,1].pie(counts, explode=explode,
             labels=labels,autopct='%1.1f%%',
             shadow=True)
ax1[0, 1].set_title('plot exploded, default startangle = 0')

#changing text size with some radius
ax1[1,0].pie(counts, explode=explode,
             labels=labels, autopct='%1.1f%%',
             shadow=True, startangle=90,
             textprops={'size': 'smaller'},
             radius = 1.5)
ax1[1, 0].set_title('changing text size with some radius')

#changing text size and radius
ax1[1,1].pie(counts, explode=explode,
             labels=labels, autopct='%1.1f%%',
             shadow=True, startangle=45,
             textprops={'size': 'x-small'}, radius = 0.75)
ax1[1,1].set_title('changing text size with some radius')

plt.show()

```

The same can be applied on the independent variables to gain some insights.
