# Pair Plot

Pair plot allows us to see the distribution of each individual variable as well as the relationship between each variable hence helping in analysis.

## Import Library for plotting

```python
# import the required libraries
import seaborn as sns;

# do not display warnings in notebook 
import warnings
warnings.filterwarnings('ignore')
```

## Load Iris dataset and Plotting

```python
sns.set(style="ticks", color_codes=True)
dataset = sns.load_dataset("iris")
plot = sns.pairplot(dataset, hue="species",size=2.0)
```

Here we can see from the distribution that setosa is easily separable whereas the other two are not.

Now that we have got a clear idea of what the data is and how it is distributed, we can apply machine learning models on it.
